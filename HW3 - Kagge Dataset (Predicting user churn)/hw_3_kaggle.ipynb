```python
X_train = data[feature_cols]
y_train = data[target_col]
```

    .

```python
param_grid = [
    {
        'logisticregression__solver': ['lbfgs'],
        'logisticregression__penalty': ['l2'],
        'logisticregression__C': [0.1, 1, 10, 100],
        'logisticregression__max_iter': [100, 200]
    },
    {
        'logisticregression__solver': ['liblinear'],
        'logisticregression__penalty': ['l1', 'l2'],
        'logisticregression__C': [0.1, 1, 10],
        'logisticregression__max_iter': [100]
    },
    {
        'logisticregression__solver': ['saga'],
        'logisticregression__penalty': ['elasticnet'],
        'logisticregression__C': [0.1, 1, 10],
        'logisticregression__l1_ratio': [0.1, 0.5],
        'logisticregression__max_iter': [200],
        'logisticregression__tol': [1e-4]
    }
]
```

     `ColumnTransformer`,         OneHot    .

```python
preprocess = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ("cat", OneHotEncoder(handle_unknown='ignore'), cat_cols)
])
```

           .

```python
pipe = make_pipeline(
    preprocess,
    LogisticRegression(random_state=42)
)
```

          .

```python
logreg = GridSearchCV(pipe,param_grid,scoring='roc_auc',cv=5,n_jobs=-1,refit=True,verbose=0)

logreg.fit(X_train, y_train)
```

```python
print(" :", logreg.best_params_)
print(f" ROC AUC: {logreg.best_score_:.4f}")
```

```python
train_data_ctb = pd.read_csv('./train.csv').replace(" ", 0)
```

       .

```python
X_train_ctb = train_data_ctb[feature_cols]
y_train_ctb = train_data_ctb[target_col]
```

  `Pool`     CatBoost.

```python
train_pool = Pool(data=X_train_ctb, label=y_train_ctb, cat_features=cat_cols)
```

  CatBoost.

```python
catboost_model = CatBoostClassifier(
    silent=True,
    eval_metric='AUC',
    random_seed=42
)
```

    .

```python
param_grid = {
    'iterations':    [200, 300],
    'depth':         [4, 6],
    'learning_rate': [0.03, 0.05],
    'l2_leaf_reg':   [1, 3],
    'subsample':     [0.7],
    'border_count':  [64],
}
```

    CatBoost.

```python
catboost_search_result = catboost_model.grid_search(
    param_grid,
    X=train_pool,
    cv=5,
    refit=True,
    verbose=False
)
```

     .

```python
best_params = catboost_search_result['params']
best_scores = catboost_search_result['cv_results']['test-AUC-mean']
max_score = max(best_scores)
```

```python
print(" :", best_params)
print(f" ROC AUC {max_score:.4f}")
```

```python
best_model = catboost_model
```

```python
X_test = pd.read_csv('./test.csv').replace(" ", 0)
test_pool = Pool(data=X_test, cat_features=cat_cols)
submission = pd.read_csv('./submission.csv')

submission['Churn'] = best_model.predict_proba(test_pool)[:, 1]
submission.to_csv('./my_submission.csv')
```
